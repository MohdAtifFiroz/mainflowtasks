```python
df = pd.read_csv("C:\\Users\\NCS\\OneDrive\\Desktop\\sql_project\\student-mat.csv" , sep=';')
```

```python
print(df.columns)
```

```python
# now average score
avg_score = df["G3"].mean()
print(" the average final score in matyhs is ", avg_score)
```

```python
# how many students scored above 15 in final grade
count_above_15 = df[df["G3"] > 15].shape[0]
print(" students score above 15 is ", count_above_15 )
```

```python
# coorelation btw study time and final grade 
correlation = df["studytime"].corr(df["G3"])
print(" the correlatin btw studytime and final grade is ",correlation )
```

```python
# gender wise average  final grade
gender_avg = df.groupby("sex")["G3"].mean()
print(" the gender wise avg is ",gender_avg)
```

```python
# plotting hostogram of final grade
plt.figure(figsize=(8,5))
plt.hist(df["G3"] , bins=10,
color='blue' , edgecolor='black')
plt.title( "distribution of final  grade" )
plt.xlabel(" final grade")
plt.ylabel(" no. of students ")
plt.show() 
```

```python
import seaborn as sns
```

```python
# scatter plot of study time and final
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x = "studytime" , y = "G3" , hue="sex")
plt.title(" study time vs final grade")
plt.xlabel(" study time ")
plt.ylabel("final grade")
plt.show()
```

```python
# bar chart for gender wise avg score
plt.figure(figsize=(6,4))
sns.barplot(x=gender_avg.index, y=gender_avg.values , palette="viridis")
plt.title(" average final grade")
plt.xlabel("gender")
plt.ylabel("avg final grade")
plt.show()
```
